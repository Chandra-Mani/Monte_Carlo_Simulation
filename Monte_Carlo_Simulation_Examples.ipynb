## Example 1: Coin Toss

 For a fair coin $P_{head} = P_{tail} = 1/2$

#### Import libraries

```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

#### defining functions

```python
def coin_toss():
    return random.randint(0,1) # 0 = head and 1 = tail
```

```python
coin_toss()
```

```python
listp=[]

def MonteCarlo(n):
    results=0
    for i in range(n):
        toss_result= coin_toss()
        results= results+toss_result
        proby=results/(i+1)
        
        listp.append(proby)
        
        #plot
        plt.axhline(y=0.5, color='k', linestyle='-')
        plt.xlabel("Iteration", fontsize=12)
        plt.ylabel("Probability", fontsize = 12)
        plt.plot(listp)
    return results/n
```

```python
answer=MonteCarlo(100)
print("final value:", answer)
```

```python
res=MonteCarlo(10000)
print("final value:", res)
```

## Example 2: Buffon's Needle Problem to estimate $\pi$

Reference: https://en.wikipedia.org/wiki/Buffon%27s_needle_problem 

Suppose we have a floor made of parallel strips of wood, each the same width, and we drop a needle onto the floor. What is the probability that the needle will lie across a line between two strips?

The needle a lies accross a line while b does not.

The probability that the needle (with $l < t$) lie in a position where it crosses one of the lines is 

$P=\tfrac{2\; \ell}{\pi\; t}$ $\implies \pi = \tfrac{2\ell}{tP}$

#### Defining function

```python
#Function to estimate PI value :
def mtCrl(runs,needles,n_length,b_width):
    #Empty list to store pi values :
    pi_values = []
    
    #Horizontal line for actual value of PI :
    plt.axhline(y=math.pi, color='g', linestyle='-')
    plt.xlabel('iterations')
    plt.ylabel('value of $\pi$')
    
    #For all runs :
    for i in range(runs):
        #Initialize number of hits as 0.
        nhits = 0
        
        #For all needles :
        for j in range(needles):
            #We will find the distance from the nearest vertical line :
            #Min = 0     Max = b_width/2
            x = random.uniform(0,b_width/2.0)
            
            #The theta value will be from 0 to pi/2 :
            theta = random.uniform(0,math.pi/2)
            
            #Checking if the needle crosses the line or not :
            xtip  = x - (n_length/2.0)*math.cos(theta)  
            if xtip < 0 :
                nhits += 1
                
        #Going with the formula :
        numerator = 2.0 * n_length * needles
        denominator = b_width * nhits
       
        #Append the final value of pi :
        pi_values.append((numerator/denominator))
    
    #Final pi value after all iterations :
    print(pi_values[-1])
    
    #Plotting the graph :
    plt.plot(pi_values)    
```

#### Call the function

```python
#Total number of runs :
runs = 1000

#Total number of needles :
needles = 100000

#Length of needle :
n_length = 2  

#space between 2 verical lines :
b_width =2

#Calling the main function :
mtCrl(runs,needles,n_length,b_width)
```

**The result is pretty close to the exact value of $\pi$.**
